```python
import pandas as pd
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

def preprocess_text(text, stop_words):
    # Preprocess the text by removing stopwords and short words
    text = ' '.join([w for w in text.split() if len(w) > 2 and w not in stop_words])
    return text

def build_model(max_length, vocabulary_size):
    model = Sequential()
    model.add(Embedding(vocabulary_size, 100, input_length=max_length))
    model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(5, activation='softmax'))
    print('max_len  :' ,max_length) 
    return model

def train_model(model, X_train, y_train, epochs=20, batch_size=32):
    # Convert labels to one-hot encoded format
    y_train_encoded = to_categorical(y_train)

    # Step 10: Compile and train the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(X_train, y_train_encoded, epochs=epochs, batch_size=batch_size)

def evaluate_model(model, X_train, y_train, X_test, y_test):
    # Step 11: Evaluate the model
    y_train_pred_prob = model.predict(X_train)
    y_test_pred_prob = model.predict(X_test)

    y_train_pred = np.argmax(y_train_pred_prob, axis=1)
    y_test_pred = np.argmax(y_test_pred_prob, axis=1)
    print('X_train shape :', X_train.shape)
    print('X_test shape :' ,X_test.shape)
    print('y_train shape :' ,y_train.shape)
    print('y_test shape :', y_test.shape)
    print(X_test)


    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    # loss, train_accuracy = model.evaluate(X_test, y_test)
    

    print('Train Accuracy:', train_accuracy)
    print('Test Accuracy:', test_accuracy)

def text_emotion_detection(dataset_path, stopwords_path):
    # Step 1: Load the dataset
    data = pd.read_csv(dataset_path)

    # Step 2: Load stopwords
    stop_words = set(open(stopwords_path, 'r').read().split())

    # Step 3: Preprocess the text
    data['Text'] = data['Text'].apply(lambda x: preprocess_text(x, stop_words))

    # Step 4: Split the dataset into train and test sets
    train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

    # Step 5: Tokenize the text
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(train_data['Text'])
    train_sequences = tokenizer.texts_to_sequences(train_data['Text'])
    test_sequences = tokenizer.texts_to_sequences(test_data['Text'])

    # Step 6: Pad sequences to have the same length
    max_length = max(len(seq) for seq in train_sequences)
    train_data_encoded = pad_sequences(train_sequences, maxlen=max_length, padding='post')
    test_data_encoded = pad_sequences(test_sequences, maxlen=max_length, padding='post')

    # Step 7: Encode emotion labels
    label_encoder = LabelEncoder()
    label_encoder.fit(train_data['Label'])
    train_labels_encoded = label_encoder.transform(train_data['Label'])
    test_labels_encoded = label_encoder.transform(test_data['Label'])

    # Step 8: Build the model
    model = build_model(max_length, len(tokenizer.word_index) + 1)

    # Step 9: Train the model
    train_model(model, train_data_encoded, train_labels_encoded)

    # Step 10: Evaluate the model
    evaluate_model(model, train_data_encoded, train_labels_encoded, test_data_encoded, test_labels_encoded)

# Run the text emotion detection system
text_emotion_detection('Dataset/Text_Emotion_Data.csv', 'Dataset/stopwords.txt')
```
