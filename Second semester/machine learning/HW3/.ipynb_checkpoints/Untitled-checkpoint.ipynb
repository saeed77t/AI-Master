```python
import random
import numpy as np
import gym
from gym import spaces


class ZombieGame(gym.Env):
    metadata = {'render.modes': ['console']}

    def __init__(self, width=10, height=10, num_zombies=3):
        self.width = width
        self.height = height
        self.num_zombies = num_zombies
        self.grid = [['.' for y in range(height)] for x in range(width)]
        self.player_position = (0, 0)
        self.zombie_positions = []
        self.generate_zombies()
        self.generate_player()
        self.action_space = spaces.Discrete(4)
        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(self.num_zombies,))

    def generate_zombies(self):
        for i in range(self.num_zombies):
            x = random.randint(0, self.width-1)
            y = random.randint(0, self.height-1)
            self.zombie_positions.append((x, y))
            self.grid[x][y] = 'Z'
    
    def generate_player(self):
        x = random.randint(0, self.width-1)
        y = random.randint(0, self.height-1)
        self.player_position = (x, y)
        self.grid[x][y] = 'P'
    
    def move_zombies(self):
        for i, (x, y) in enumerate(self.zombie_positions):
            dx, dy = self.get_zombie_direction(x, y)
            new_x, new_y = x+dx, y+dy
            if self.is_valid_position((new_x, new_y)):
                self.grid[x][y] = '.'
                self.zombie_positions[i] = (new_x, new_y)
                self.grid[new_x][new_y] = 'Z'
    
    def get_zombie_direction(self, x, y):
        px, py = self.player_position
        dx = px - x
        dy = py - y
        if abs(dx) > abs(dy):
            return (dx//abs(dx), 0)
        else:
            return (0, dy//abs(dy))
    
    def move_player(self, action):
        x, y = self.player_position
        if action == 0:
            new_x, new_y = x-1, y
        elif action == 1:
            new_x, new_y = x+1, y
        elif action == 2:
            new_x, new_y = x, y-1
        elif action == 3:
            new_x, new_y = x, y+1
        else:
            raise ValueError(f"Invalid action: {action}")
        
        if self.is_valid_position((new_x, new_y)):
            self.grid[x][y] = '.'
            self.player_position = (new_x, new_y)
            self.grid[new_x][new_y] = 'P'
    
    def is_valid_position(self, position):
        x, y = position
        return 0 <= x < self.width and 0 <= y < self.height and self.grid[x][y] != 'X'
    
    def get_state(self):
        # Get the positions of the zombies and the player
        player_x, player_y = self.player_position
        zombie_positions = self.zombie_positions

        # Compute the distances from the player to each zombie
        distances = [np.sqrt((player_x - zombie_x) ** 2 + (player_y - zombie_y) ** 2)
                     for zombie_x, zombie_y in zombie_positions]

        # Return a tuple of the distances
        return tuple(distances)

    
    def reset(self):
        # Reset the state of the game
        self.grid = [['.' for y in range(self.height)] for x in range(self.width)]
        self.player_position = (0, 0)
        self.zombie_positions = []
        self.generate_zombies()
        self.generate_player()

        # Return the initial state
        return self.get_state()

    def step(self, action):
        # Move the player in the chosen direction
        self.move_player(action)

        # Move each zombie towards the player
        self.move_zombies()

        # Compute the reward and check if the game is done
        reward = self.compute_reward()
        done = self.check_game_over()

        # Compute the new state
        state = self.get_state()

        # Return the new state, reward, and done flag
        return state, reward, done, {}

    def render(self, mode='console'):
        # Render the game in the console
        if mode != 'console':
            raise NotImplementedError()

        # Print the game grid
        for row in self.grid:
            print(' '.join(row))
        print()

    def is_valid_position(self, position):
        x, y = position
        return 0 <= x < self.width and 0 <= y < self.height and self.grid[x][y] != 'X'

    def move_player(self, action):
        x, y = self.player_position
        if action == 0:
            new_x, new_y = x-1, y
        elif action == 1:
            new_x, new_y = x+1, y
        elif action == 2:
            new_x, new_y = x, y-1
        elif action == 3:
            new_x, new_y = x, y+1
        else:
            raise ValueError(f"Invalid action: {action}")
        
        if self.is_valid_position((new_x, new_y)):
            self.grid[x][y] = '.'
            self.player_position = (new_x, new_y)
            self.grid[new_x][new_y] = 'P'

    def get_zombie_direction(self, x, y):
        px, py = self.player_position
        dx = px - x
        dy = py - y
        if abs(dx) > abs(dy):
            return (dx//abs(dx), 0)
        else:
            return (0, dy//abs(dy))

    def move_zombies(self):
        for i, (x, y) in enumerate(self.zombie_positions):
            dx, dy = self.get_zombie_direction(x, y)
            new_x, new_y = x+dx, y+dy
            if self.is_valid_position((new_x, new_y)):
                self.grid[x][y] = '.'
                self.zombie_positions[i] = (new_x, new_y)
                self.grid[new_x][new_y] = 'Z'

    def generate_zombies(self):
        for i in range(self.num_zombies):
            x = random.randint(0, self.width-1)
            y = random.randint(0, self.height-1)
            self.zombie_positions.append((x, y))
            self.grid[x][y] = 'Z'

    def generate_player(self):
        x = random.randint(0, self.width-1)
        y = random.randint(0, self.height-1)
        self.player_position = (x, y)
        self.grid[x][y] = 'P'

    def compute_reward(self):
        # Get the distances to each zombie
        distances = self.get_state()

        # If the player is within one step of a zombie, return -1
        if any(distance <= 1 for distance in distances):
            return -1

        # If the player is farther than one step from all zombies, return 0
        if all(distance > 1 for distance in distances):
            return 0

        # If the player is not adjacent to any zombies but is closer to some than others, return 1/n for each zombie that
        # is the closest to the player, where n is the number of closest zombies
        closest_distance = min(distance for distance in distances if distance > 1)
        closest_zombies = [i for i, distance in enumerate(distances) if distance == closest_distance]
        return 1 / len(closest_zombies)

    def check_game_over(self):
        # Check if the game is over
        player_x, player_y = self.player_position

        # If the player is adjacent to a zombie, the game is over
        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                if dx == 0 and dy == 0:
                    continue
                x = player_x + dx
                y = player_y + dy
                if self.is_valid_position((x, y)) and self.grid[x][y] == 'Z':
                    return True

        # If the player is not adjacent to any zombies but there are no valid moves left, the game is over
        if all(not self.is_valid_position((player_x+dx, player_y+dy)) for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]):
            return True

        # Otherwise, the game is not over
        return False
```

```python
import numpy as np
import random
from collections import defaultdict
import math

class PacManRL:
    def __init__(self, num_episodes=1000, gamma=0.99, alpha=0.5, epsilon=1.0, epsilon_decay=0.01, epsilon_min=0.01):
        self.q_table = {}
        self.num_episodes = num_episodes
        self.gamma = gamma
        self.alpha = alpha
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        # Define the action space and initialize the Q-table
        self.actions = [0,1,2,3]
        self.q_table = defaultdict(lambda: np.zeros(len(self.actions)))

        # Define features for state representation
        self.food_left = None
        self.food_eaten = None

    def reset(self):
        self.food_left = None
        self.food_eaten = None
        
    def get_state(self, game_state):
        # Extract relevant information from the game state to represent it as a state
        if isinstance(game_state, dict):
            pacman_pos = game_state.get('pacman_pos', ())
            food_pos = tuple(map(tuple, np.argwhere(game_state['food'])))
            ghost_pos = tuple(map(tuple, game_state['ghost_pos']))
        else:
            pacman_pos = ()
            food_pos = ()
            ghost_pos = ()

        # Compute the remaining food left
        if self.food_left is None:
            self.food_left = len(food_pos)
        else:
            self.food_left -= self.food_eaten

        # Compute the distance to the nearest food and the nearest ghost
        dist_to_food = np.inf
        for pos in food_pos:
            dist = np.linalg.norm(np.array(pacman_pos) - np.array(pos))
            if dist < dist_to_food:
                dist_to_food = dist

        dist_to_ghost = np.inf
        for pos in ghost_pos:
            dist = np.linalg.norm(np.array(pacman_pos) - np.array(pos))
            if dist < dist_to_ghost:
                dist_to_ghost = dist

        # Compute the new features
        self.food_eaten = self.food_left - len(food_pos)
        features = {
            'bias': 1.0,
            'food_distance': dist_to_food,
            'ghost_distance': dist_to_ghost
        }

        return {'pacman_pos': pacman_pos, 'food_pos': food_pos, 'ghost_pos': ghost_pos}
    
    
    
    
    
    
    
    
    
    
    
    

    def get_epsilon(self, episode):
        """
        Returns the exploration rate for the given episode.
        """
        return max(0.01, min(1, 1.0 - math.log10((episode + 1) / 25)))


    def choose_action(self, state, epsilon):
        # Choose an action using an epsilon-greedy policy
        if np.random.uniform(0, 1) < epsilon:
            action = np.random.choice(self.actions)
        else:
            q_values = self.q_table.get(state, {a: 0 for a in self.actions})
            max_q_value = max(q_values.values())
            actions_with_max_q_value = [a for a, q in q_values.items() if q == max_q_value]
            action = np.random.choice(actions_with_max_q_value)
        return action

    def update_q(self, state, action, next_state, reward, alpha, gamma):
        # Update Q-value for the given state-action pair
        td_target = reward + gamma * np.max(self.q_table[next_state])
        td_error = td_target - self.q_table[state][self.actions.index(action)]
        self.q_table[state][self.actions.index(action)] += alpha * td_error

    def run_episode(self, env):
        # Reset the environment and initialize the state
        obs = env.reset()
        state = self.get_state(obs)

        done = False
        total_reward = 0

        while not done:
            # Choose an action and take a step in the environment
            action = self.choose_action(state, self.epsilon)
            obs, reward, done, info = env.step(action)
            total_reward += reward

            # Update the state and Q-table
            next_state = self.get_state(obs)
            self.update_q(state, action, next_state, reward, self.alpha, self.gamma)

            state = next_state

        return total_reward

    def train(self, env):
        # Initialize the Q-table
        self.q_table = {}

        for i in range(self.num_episodes):
            obs = env.reset()
            state = self.get_state(obs)
            done = False
            total_reward = 0
            epsilon = self.get_epsilon(i)

            while not done:
                # Choose an action and take a step in the environment
                action = self.choose_action(state, epsilon)
                obs, reward, done, info = env.step(action)
                total_reward += reward

                # Update the Q-table
                next_state = self.get_state(obs)
                self.update_q(state, action, next_state, reward, self.alpha, self.gamma)
                state = next_state

            # Update epsilon
            if self.epsilon > self.epsilon_min:
                self.epsilon -= self.epsilon_decay

        return self.q_table
```
