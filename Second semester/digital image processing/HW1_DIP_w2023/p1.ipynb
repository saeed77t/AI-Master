```python
import cv2
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('imgs/barbara.png')

# Image negative
s_img = 255 - 1 - img

plt.imshow(img)
plt.show()
plt.imshow(s_img)
plt.show()
```

```python
# log transformation
c = [1, 2, 3]

for c_i in c:
    s_img = c_i * np.log10(1 + img)
    s_img = np.divide(s_img, 255)
    # Fixing out of bounds pixels
    s_img[s_img < 0] = 0
    s_img[s_img > 1] = 1
    
    plt.imshow(s_img.tolist())
    plt.show()
```

```python
# gamma correction
c = 1
gammas = [0.4, 1.0, 1.5]

for gamma in gammas:
    s_img = np.array(c * np.power(img, gamma))
    s_img = np.divide(s_img, 255)
    # Fixing out of bounds pixels
    s_img[s_img < 0] = 0
    s_img[s_img > 1] = 1
    
    plt.imshow(s_img.tolist())
    plt.show()
```

```python
K_s = [45, 128, 225]

for k in K_s:
    s_img = np.array(img)
    s_img[s_img >= k] = 255
    s_img[s_img < k] = 0
    s_img = np.divide(s_img, 255)
    # Fixing out of bounds pixels
    s_img[s_img < 0] = 0
    s_img[s_img > 1] = 1
    
    plt.imshow(s_img.tolist())
    plt.show()
```
