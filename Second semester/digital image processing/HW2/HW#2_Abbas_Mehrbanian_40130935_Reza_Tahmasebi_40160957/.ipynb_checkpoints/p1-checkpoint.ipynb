```python
import cv2
import numpy as np
import utils
from scipy.fftpack import fft2, ifft2, fftshift
import matplotlib.pyplot as plt
```

## Part I

### Reading Images

```python
from os import listdir

test_images = []
for images in listdir("test images"):
    if (images.endswith(".bmp")):
        test_images.append(images)

test_images.sort()
```

### Applying Fourier Transform

The Fourier transform of an image decomposes it into a series of sinusoidal signals of different frequencies. The magnitude of the Fourier transform, also known as the magnitude spectrum, represents the strength of these frequencies.

When examining the magnitude spectrum of an image, noise or other patterns in the image can be identified by examining the frequency content at different spatial frequencies. For example, high-frequency content represents fine details and edges in the image, while low-frequency content represents smooth variations in the image.

If there is noise in the image, it will often appear as high-frequency content in the magnitude spectrum. This is because noise introduces random variations in the image that are not correlated with the underlying image content. By examining the high-frequency content of the magnitude spectrum, it is possible to identify the presence of noise and the type of noise that is present in the image.

Similarly, other patterns in the image can also be identified by examining their frequency content in the magnitude spectrum. For example, periodic patterns in the image will often appear as peaks in the magnitude spectrum at specific frequencies. By examining the frequency content of the magnitude spectrum, it is possible to gain insights into the underlying structure and content of the image.

```python
for im_name in test_images:
    # Load the image using OpenCV
    im = utils.normalize_image(cv2.imread(f"test images/{im_name}", cv2.IMREAD_GRAYSCALE))

    # Apply the Fourier transform
    im_fft = fft2(im)

    # Shift the spectrum using fftshift
    im_fft_shifted = fftshift(im_fft)

    # Create a figure with three subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))
    fig.suptitle(f"{im_name}", fontsize=12, fontweight="bold")

    # Plot the original image
    axs[0].imshow(im, cmap='gray')
    axs[0].set_title('Original Image', fontsize=10)
    axs[0].axis('off')

    # cv2.imshow(im_name, im)
    # cv2.waitKey(0)

    # Plot the Fourier transform
    magnitude_spectrum = 20 * np.log(1 + utils.abs(im_fft_shifted))
    axs[1].imshow(magnitude_spectrum, cmap='gray')
    axs[1].set_title('Fourier Transform', fontsize=10)
    axs[1].axis('off')

    # Show the plot
    plt.tight_layout()

    fig = plt.figure(figsize=(4, 4))
    ax = fig.add_subplot(projection='3d')
    X, Y = np.meshgrid(range(magnitude_spectrum.shape[1]), range(magnitude_spectrum.shape[0]))
    ax.plot_surface(X, Y, magnitude_spectrum, cmap='viridis', linewidth=0)
    # ax.axis('off')
    plt.tight_layout()
```

#### Images analysis

##### img_t1.bmp

The manitude spectrum is only showing a dot, it could mean that the initial image was made with only a costant single point repeating pattern or given this Sin formula below:

$$ y = A  Sin(\frac{2\pi x}{\lambda }+\phi ) $$

A was equal to zero. A is the amplitude of the wave, which determines how high and low the wave goes.

##### img_t3.bmp & img_t4.bmp

Since the magnitude spectrum of fourier transform is also showing a line we concluded that the intial image was made with a linear constant function. (Like X = 2)

The direction of this function (X or Y axis) can also be understood from spectrum plots.

##### img_t5.bmp & img_t6.bmp

These images are made by a Sin/Cos function. They could be made from (X * sinc(uXpi)) as well but according to spectrum plots we concluded that the initial image before fourier transform was made with a Sin/os function.

The spectrum magnitude also shows us the direction (X or Y axis) of the Sin/Cos function.

```python
hW, hH = 300, 300
hFreq = 20

x = np.linspace(0, 2*hW/(2*hW +1), 2*hW+1)     # columns (Width)
y = np.linspace(0, 2*hH/(2*hH +1), 2*hH+1)     # rows (Height)
[X,Y] = np.meshgrid(x,y)
```

```python
A = np.cos(hFreq*np.pi*X)

plt.imshow(A, cmap = 'gray');
plt.axis('off')
plt.show()
```

```python
A = np.cos(hFreq*np.pi*Y)

plt.imshow(A, cmap = 'gray');
plt.axis('off')
plt.show()
```

##### img_t7.bmp

This initial image pattern before Fourier transform is obtained by multiplying the above two Sin or Cos functions since there are multiple peaks in spectrum plot. The same thing that we said about Sinc in last part applies to this part as well.

```python
hFreq = 20
vFreq = 20

A1 = (np.cos(hFreq*np.pi*X)) * (np.cos(vFreq*np.pi*Y))

plt.figure()
plt.imshow(A1, cmap = 'gray')
plt.axis('off')
plt.show()
```

##### img_t10.bmp

The image is made with a rectangle centred at origin with sides of length X and Y. consider a function f(m,n) that equals 1 within a rectangular region and 0 everywhere else. The shape of its Fourier transform can be described as a combination of Sinc functions:

$$ F(u, v) = XY Sinc(\pi Xu) Sinc(\pi Yv) $$

The peak at the center of the plot is F(0,0), which is the sum of all the values in f(m,n). The 3D spectrum of t10 shows that F(ω1,ω2) has more energy at high horizontal frequencies than at high vertical frequencies. This reflects the fact that horizontal cross sections of f(m,n) are narrow pulses, while vertical cross sections are broad pulses. Narrow pulses have more high-frequency content than broad pulses.

##### img_t8.bmp & img_t9.bmp

We analysed t10 first so we could explain these two images with more ease. The logic is true about these two image but for t8 the energy at high horizontal frequencies are maximum and for t9 the the energy at high vertical frequencies are maximum. 

The Fourier transform is also seems to be a sinc funcion in form of:

$$ F(u, v) = X Sinc(\pi Xu) $$

for t8 Fourier transform and for t9 Fourier transform:
