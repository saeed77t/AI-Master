```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os

def edge_detection(image, method, threshold1, threshold2, L2gradient=False):
    # Rest of the code remains the same

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply the chosen edge detection method
    if method == 'canny':
        edges = cv2.Canny(gray_image, threshold1, threshold2)
    elif method == 'sobel':
        sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
        edges = np.sqrt(sobelx ** 2 + sobely ** 2).astype(np.uint8)
    elif method == 'laplacian':
        laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
        edges = cv2.convertScaleAbs(laplacian)
        _, edges = cv2.threshold(edges, threshold1, threshold2, cv2.THRESH_BINARY)
    elif method == 'deep':
        # Load the DeepLab model
        img = image
        (H, W) = img.shape[:2]
        blob = cv2.dnn.blobFromImage(img, scalefactor=1.0, size=(W, H),swapRB=True, crop=False)
        net = cv2.dnn.readNetFromCaffe("/kaggle/input/hw3dip/deploy.prototxt", "/kaggle/input/hw3dip/hed_pretrained_bsds.caffemodel")
        net.setInput(blob)
        hed = net.forward()
        hed = cv2.resize(hed[0, 0], (W, H))
        hed = (255 * hed).astype("uint8")

#         _,hed = cv2.threshold(hed, 100,255, cv2.THRESH_BINARY)
        # Threshold the edge map to obtain binary edges
        _, edges = cv2.threshold(hed, threshold1,threshold2, cv2.THRESH_BINARY)
    else:
        raise ValueError("Invalid edge detection method.")

    return edges










def hough_transform(image, threshold ,minLineLength ,maxLineGap):
    # Convert the image to uint8
    image = image.astype(np.uint8)
    # Perform the Hough transform
#     lines = cv2.HoughLines(image, 1, np.pi / 180, threshold=threshold)
        # Perform the Hough transform
    
    
#     # Print the rho and theta values for each line
#     if lines is not None:
#         for line in lines:
#             rho, theta = line[0]
#             print("Rho:", rho, "Theta:", theta)
    
    lines = cv2.HoughLinesP(image, 1, np.pi / 180, threshold=threshold ,minLineLength=minLineLength, maxLineGap=maxLineGap)
    
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            print("Line Segment: (", x1, ",", y1, ")-(", x2, ",", y2, ")")
  

    return lines




# def hough_transform2(edges, threshold):
#     # Perform the Hough transform
#     lines = cv2.HoughLines(edges, 1, np.pi / 180, threshold=threshold)

#     # Print the rho and theta values for each line
#     if lines is not None:
#         for line in lines:
#             rho, theta = line[0]
#             print("Rho:", rho, "Theta:", theta)
            
            
def hough_transform2(edges, threshold):
    # Apply thresholding to convert the edges to a binary image
    _, edges_binary = cv2.threshold(edges, 0, 255, cv2.THRESH_BINARY)

    # Perform the Hough transform
    lines = cv2.HoughLines(edges_binary, 1, np.pi / 180, threshold=threshold)

    # Print the rho and theta values for each line
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            print("Rho:", rho, "Theta:", theta)

    return lines







def plot_hough_lines(image, edges, lines):
    # Create subplots for original image, edges, and lines
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    
    # Plot original image
    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Original Image")
    
    # Plot edges
    axes[1].imshow(edges, cmap='gray')
    axes[1].set_title("Edges")
    
    # Create a copy of the original image
    image_copy = np.copy(image)
    
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(image_copy, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    # Convert BGR to RGB for matplotlib
    image_rgb = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
    
    # Plot lines
    axes[2].imshow(image_rgb)
    axes[2].set_title("Lines")
    
    # Set spacing between subplots
    plt.tight_layout()
    
    # Show the plot
    plt.show()
    
    
    
    
    
    
def plot_hough_lines2(image, edges, lines):
    # Create subplots for original image, edges, and lines
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    
    # Plot original image
    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Original Image")
    
    # Plot edges
    axes[1].imshow(edges, cmap='gray')
    axes[1].set_title("Edges")
    
    # Create a copy of the original image
    image_copy = np.copy(image)
    
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(image_copy, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    # Convert BGR to RGB for matplotlib
    image_rgb = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
    
    # Plot lines
    axes[2].imshow(image_rgb)
    axes[2].set_title("Lines")
    
    # Set spacing between subplots
    plt.tight_layout()
    
    # Show the plot
    plt.show()
    
    
    
def hough_circle_detection(image, method, dp, minDist, param1, param2, minRadius, maxRadius):
    # Apply edge detection
    edges = edge_detection(image, method, threshold1=100, threshold2=255)

    # Perform Hough circle detection
    circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=dp, minDist=minDist,
                               param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)

    return circles, edges


def plot_circle_detection(image, edges, circles):
    # Create subplots for the original image, edges, and detected circles
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    # Plot the original image
    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Original Image")

    # Plot the edges
    axes[1].imshow(edges, cmap='gray')
    axes[1].set_title("Edges")

    # Create a copy of the original image
    image_copy = np.copy(image)

    if circles is not None:
        circles = np.round(circles[0, :]).astype(int)
        for (x, y, r) in circles:
            # Draw the detected circles on the image
            cv2.circle(image_copy, (x, y), r, (0, 255, 0), 2)

    # Plot the image with detected circles
    axes[2].imshow(cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB))
    axes[2].set_title("Detected Circles")

    # Set spacing between subplots
    plt.tight_layout()

    # Show the plot
    plt.show()

    
    
def detect_and_draw_circles(image, dp, minDist, param1, param2, minRadius, maxRadius):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Hough circle detection
    circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, dp, minDist, param1=param1,
                               param2=param2, minRadius=minRadius, maxRadius=maxRadius)

    if circles is not None:
        circles = np.round(circles[0, :]).astype(int)
        for (x, y, r) in circles:
            # Draw a bounding box around the circle
            x1 = x - r
            y1 = y - r
            x2 = x + r
            y2 = y + r
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return image




```

# functions
### 1. edge_detection(image, method, threshold1, threshold2, L2gradient=False):
This function applies edge detection to the input image using the specified method. It supports four edge detection methods: Canny, Sobel, Laplacian, and Deep. The function converts the image to grayscale and applies the selected method to obtain the edges.
### 2. hough_transform(image, threshold, minLineLength, maxLineGap):
This function performs the Hough transform for line detection on the input image. It takes the threshold, minimum line length, and maximum line gap as parameters and returns the detected lines in the image using the HoughLinesP function.
The Hough transform is a technique used to detect simple shapes, such as lines or circles, in an image. It works by converting the points in the image space to parameter space, where each shape corresponds to a curve. In the case of the Hough transform for lines, the parameter space is represented by the rho (ρ) and theta (θ) values, which define a line in polar coordinates. The steps involved in the Hough transform for line detection are as follows:
1. Convert the image to grayscale.
2. Apply edge detection to obtain a binary image with edge pixels.
3. Initialize an accumulator matrix in the parameter space, where each element corresponds to a line in the image space.
4. For each edge pixel in the binary image, iterate over a range of theta values and calculate the corresponding rho value using the equation: ρ = x * cos(θ) + y * sin(θ), where (x, y) are the coordinates of the edge pixel.
5. Increment the accumulator matrix at the corresponding (ρ, θ) position.
6. Threshold the accumulator matrix to obtain the lines with sufficient votes (indicating their presence in the image).
7. Interpret the lines from the accumulator matrix by extracting the rho and theta values and converting them back to the image space.
8. plot_hough_lines(image, edges, lines): This function plots the original image, edges, and lines detected by the Hough transform. It visualizes the image and draws the detected lines on a copy of the original image.
###  3. hough_transform2(edges, threshold):
This function performs the Hough transform for line detection on the input edge image. It applies thresholding to convert the edges to a binary image and uses the HoughLines function to detect lines. It prints the rho and theta values for each detected line.
The Hough transform for line detection using the HoughLines function differs slightly from the HoughLinesP function. Instead of representing lines as line segments, the HoughLines function represents them as infinite lines passing through each (ρ, θ) point in the parameter space.
### 4. plot_hough_lines2(image, edges, lines):
This function plots the original image, edges, and lines detected by the Hough transform. It visualizes the image and draws the detected lines on a copy of the original image using the rho and theta values.
### 5. hough_circle_detection(image, method, dp, minDist, param1, param2, minRadius, maxRadius):
This function performs the Hough transform for circle detection on the input image. It applies edge detection using the specified method, and then applies the HoughCircles function to detect circles. It returns the detected circles and the edges.
The Hough transform for circle detection is similar to the one for line detection, but with different parameterization. Instead of rho and theta, circles are parameterized by their center coordinates (x, y) and radius (r). The HoughCircles function accumulates votes for potential circle centers and radii in the parameter space, and the peaks in the accumulator matrix correspond to the detected circles.
### 6. plot_circle_detection(image, edges, circles):
This function plots the original image, edges, and circles detected by the Hough transform. It visualizes the image and draws the detected circles on a copy of the original image.
### 7. detect_and_draw_circles(image, dp, minDist, param1, param2, minRadius, maxRadius): 
This function detects circles in the given image using the Hough circle transform and draws bounding boxes around the circles. It converts the image to grayscale, applies Gaussian blur to reduce noise, and then applies the HoughCircles function. It draws bounding boxes around the detected circles and returns the modified image.
Overall, this code provides functions for edge detection, line detection using the Hough transform, and circle detection using the Hough circle transform. It also includes functions to visualize the results. The Hough transform is a powerful technique for shape detection in images, allowing the identification of lines and circles even in the presence of noise and partial occlusion.
