```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os

# def edge_detection(image, method, threshold1, threshold2):
#     # Convert the image to grayscale
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
#     # Apply the chosen edge detection method
#     if method == 'canny':
#         edges = cv2.Canny(gray_image, threshold1, threshold2)
#     elif method == 'sobel':
#         sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
#         sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
#         edges = np.sqrt(sobelx**2 + sobely**2).astype(np.uint8)
#     else:
#         raise ValueError("Invalid edge detection method.")
    
#     return edges

# def edge_detection(image, method, threshold1, threshold2):
#     # Convert the image to grayscale
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
#     # Apply the chosen edge detection method
#     if method == 'canny':
#         edges = cv2.Canny(gray_image, threshold1, threshold2)
#     elif method == 'sobel':
#         sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
#         sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
#         edges = np.sqrt(sobelx**2 + sobely**2).astype(np.uint8)
#     elif method == 'probabilistic_hough':
#         # Apply probabilistic Hough transform for line detection
#         edges = cv2.Canny(gray_image, threshold1, threshold2)
#         lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=50, minLineLength=50, maxLineGap=10)
#         return lines
#     else:
#         raise ValueError("Invalid edge detection method.")
    
#     return edges.astype(np.uint8)  # Convert the edges to the correct data type





# def edge_detection(image, method, threshold1, threshold2):
#     # Convert the image to grayscale
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
#     # Apply the chosen edge detection method
#     if method == 'canny':
#         edges = cv2.Canny(gray_image, threshold1, threshold2)
#     elif method == 'sobel':
#         sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
#         sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
#         edges = np.sqrt(sobelx**2 + sobely**2).astype(np.uint8)
#     elif method == 'laplacian':
#         laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
#         edges = cv2.convertScaleAbs(laplacian)
#         _, edges = cv2.threshold(edges, threshold1, threshold2, cv2.THRESH_BINARY)
#     else:
#         raise ValueError("Invalid edge detection method.")
    
#     return edges


def edge_detection(image, method, threshold1=100, threshold2,200 , L2gradient=False):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply the chosen edge detection method
    if method == 'canny':
        edges = cv2.Canny(gray_image, threshold1, threshold2 , L2gradient = L2gradient)
    elif method == 'sobel':
        sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
        edges = np.sqrt(sobelx ** 2 + sobely ** 2).astype(np.uint8)
    elif method == 'laplacian':
        laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
        edges = cv2.convertScaleAbs(laplacian)
        _, edges = cv2.threshold(edges, threshold1, threshold2, cv2.THRESH_BINARY)
    elif method == 'deep':
        # Load the DeepLab model
        model_path = "deeplabv3_mnv2_pascal_train_aug.pb"
        net = cv2.dnn.readNet(model_path)

        # Prepare the input blob
        blob = cv2.dnn.blobFromImage(image, scalefactor=1.0, size=(513, 513), mean=(127.5, 127.5, 127.5), swapRB=True)

        # Set the input blob for the network
        net.setInput(blob)

        # Forward pass through the network to get the edge predictions
        predictions = net.forward()

        # Extract the edge map from predictions
        edge_map = predictions[0, 1, :, :]

        # Threshold the edge map to obtain binary edges
        _, edges = cv2.threshold(edge_map, 0.5, 1, cv2.THRESH_BINARY)

        # Convert the edge map to uint8 format
        edges = (edges * 255).astype(np.uint8)
    else:
        raise ValueError("Invalid edge detection method.")

    return edges






# def edge_detection(image, method, threshold1, threshold2):
#     # Convert the image to grayscale
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#     # Apply the chosen edge detection method
#     if method == 'canny':
#         edges = cv2.Canny(gray_image, threshold1, threshold2)
#     elif method == 'sobel':
#         sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
#         sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
#         edges = np.sqrt(sobelx ** 2 + sobely ** 2).astype(np.uint8)
#     elif method == 'laplacian':
#         laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
#         edges = cv2.convertScaleAbs(laplacian)
#         _, edges = cv2.threshold(edges, threshold1, threshold2, cv2.THRESH_BINARY)
#     elif method == 'deep':
#         # Load the HED model
#         model_path = "hed_pretrained_model.pb"
#         if not os.path.isfile(model_path):
#             urllib.request.urlretrieve(
#                 "https://raw.githubusercontent.com/s9xie/hed/master/examples/hed/HED_pretrained_bsds.caffemodel",
#                 model_path)
#         proto_path = "deploy.prototxt"
#         if not os.path.isfile(proto_path):
#             urllib.request.urlretrieve(
#                 "https://raw.githubusercontent.com/s9xie/hed/master/examples/hed/deploy.prototxt",
#                 proto_path)
#         net = cv2.dnn.readNet(proto_path, model_path)

#         # Prepare the input blob
#         blob = cv2.dnn.blobFromImage(image, scalefactor=1.0, size=(256, 256),
#                                      mean=(104.00698793, 116.66876762, 122.67891434),
#                                      swapRB=True, crop=False)

#         # Set the input blob for the network
#         net.setInput(blob)

#         # Forward pass through the network to get the edge predictions
#         predictions = net.forward()

#         # Reshape the predictions to get the edge map
#         edge_map = predictions[0, 0, :, :]
#         edge_map[edge_map < 128] = 0
#         edge_map[edge_map >= 128] = 255

#         # Convert the edge map to uint8 format
#         edges = edge_map.astype(np.uint8)
#     else:
#         raise ValueError("Invalid edge detection method.")

#     return edges








def hough_transform(image, threshold):
    # Convert the image to uint8
    image = image.astype(np.uint8)
    # Perform the Hough transform
    lines = cv2.HoughLines(image, 1, np.pi / 180, threshold=threshold)
    
    return lines



# def hough_transform2(image, threshold):
#     # Convert the image to uint8
#     image = image.astype(np.uint8)
#     # Perform the Hough transform
#     lines = cv2.HoughLines(image, 1, np.pi / 180, threshold=threshold)
    
#     return lines




def plot_hough_lines(image, edges, lines):
    # Create subplots for original image, edges, and lines
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    
    # Plot original image
    axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Original Image")
    
    # Plot edges
    axes[1].imshow(edges, cmap='gray')
    axes[1].set_title("Edges")
    
    # Create a copy of the original image
    image_copy = np.copy(image)
    
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(image_copy, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    # Convert BGR to RGB for matplotlib
    image_rgb = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
    
    # Plot lines
    axes[2].imshow(image_rgb)
    axes[2].set_title("Lines")
    
    # Set spacing between subplots
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Load the images using cv2
bldg_image = cv2.imread("bldg.png")
greek_image = cv2.imread("Greek.jpg")

# Set the edge detection method ('canny' or 'sobel')
edge_method = 'canny'
# edge_method = 'sobel'

# Set the threshold values for edge detection
threshold1 = 15
threshold2 = 20
houghthreshold = 200
# Apply edge detection on the images
bldg_edges = edge_detection(bldg_image, 'laplacian', threshold1, threshold2)
greek_edges = edge_detection(greek_image, 'canny', 199, 280)

# Apply Hough transform on the images
bldg_lines = hough_transform(bldg_edges, houghthreshold)
greek_lines = hough_transform(greek_edges,99)

# Plot the original images, edges, and lines
plot_hough_lines(bldg_image, bldg_edges, bldg_lines)
plot_hough_lines(greek_image, greek_edges, greek_lines)
```
