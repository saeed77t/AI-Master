```python
#importing libraries 
import random as rand
import numpy as np
```

```python
#Defining Values 

population_size = 100
mutation_probability = 0.8
number_of_queens = 8
```

```python
def InitialPopulation(popsize , n):
    #popsize is population size and n is numberof qeens
    generations = []
    for i in range(popsize):
        # creating  first population with n chromosom
        a =rand.sample(range(1, n+1) , n)
       # a  =  a.tolist()
        #adding fittnes number add the end of the chromosom
        a.append(0)
        generations.append(a)
        
    return generations    
```

```python
#sorting population based on the fittness , fitness is the the last member of choromosoms
def FitSort(population , n):
    for i in range(len(population)):
        min = i
        for j in range(i,len(population)):
            if population[j][n]<population[min][n]:
                min = j
        temp =  population[i]
        population[i] = population[min]
        population[min] = temp 
        return population

    
```

```python
#calculating fitness : best fitness is no queens can check on each other , so best fitness here is 0 
def Fitness(population , n ):
    check = 0
    for i in range(len(population)):
        check = 0 
        for j in range(n):
            for b in range(j):
                #this line here is usless beacuse our population is reperesented permuted so we dont have repetitious members
                if population[i][j] == population[i][b]:
                    check += 1
                    #this condition calculates if queens are cheking on each other diagonal ! 
                if abs(j-b)==abs(population[i][j]-population[i][b]): 
                    check+=1
                    
        population[i][n] = check
        
    # soting population based on fittness    
    population =FitSort(population , n)   
    
    return population

            
            
            
            
         
```

```python
#selecting parents out of population
def ParentSelection(population , m):
    n = len(population)
    #choosing random  parents among the population
    parent = rand.sample(range(0 , n),5)
    a = []
    #getting the 5 random out of population and sorting them,
    for i in range(len(parent)):
        b = population[parent[i]][m] , parent[i]
        a.append(b)
        a.sort()
        # choosing top 2 out of 5
    parent1 = population[a[0][1]]
    parent2 = population[a[1][1]]
    return parent1 , parent2
```

```python
# one point cross over ,
def CrossOver(parent1 , parent2):
    # defining 4 genes out of 2 parents 
    gene1=[]
    gene2 = []
    gene3 = []
    gene4 = []
    child1=[]
    child2=[]
    #breaking parent 1 into two genes , gene1 and gene2
    for i in range(len(parent1) -1):
        if i < (len(parent1) -1) / 2:
            gene1.append(parent1[i])
        else:
            gene2.append(parent1[i])
            #breaking  parent2   into 2 genes , gene3 and gene4
    for i in range(len(parent2) -1):
        if i < (len(parent2) -1) / 2:
            gene3.append(parent2[i])
        else:
            gene4.append(parent2[i])        
   # print("gene1 = ",gene1)
  #  print("gene2 = ", gene2)
  #  print("gene3 = ",gene3)
  #  print("gene4 = ", gene4)
    
    #crossing over the genes !
    child1 = gene1+gene4
    child2 = gene2 + gene3

    
   # print("child1 = " ,child1)
  #  print("child2 = " ,child2)
    
    return child1 , child2
        
        
    
```

```python
#swap mutation
def Mutation(child , mutatonRate):
    # choosing a random number between 1 to 100
    m = rand.randint(1,100)
    if m < mutatonRate:
        #choosing two chromosoms to swap
        gene_index= rand.sample(range(0 , len(child)-1),2)
        temp1 = child[gene_index[0]]
        temp2 = child[gene_index[1]]
        child[gene_index[1]] = temp1
        child[gene_index[0]] = temp2
        child.append(0)
        
    else:
        child.append(0)
        
    return child   
        
    
```

```python
# survival selection loop
def SurvivalSelection(population , childs , n):
    #adding childs to population
    for i in range(len(childs)):
        population.append(childs[i])
    #sorting population
    population = FitSort(population , n)
    m = len(population)
    #ommiting 2 worst out of population
    del population[m-1]
    del population[m-2]
    
    

    
    return population
        
    
```
